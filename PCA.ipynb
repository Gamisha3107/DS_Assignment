2. Dimensionality Reduction with PCA

```python
from sklearn.preprocessing import StandardScaler

x = df.drop('Alcohol', axis=1)
y = df['Alcohol']

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.decomposition import PCA

pca = PCA()
x_pca = pca.fit_transform(x_scaled)

explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)

plt.figure(figsize=(10,6))
plt.plot(range(1, len(explained_variance_ratio)+1), explained_variance_ratio, marker='o', label='Individual Variance')
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='s', label='Cumulative Variance', color='yellow')
plt.axhline(y=0.95, color='orange', linestyle='--', label='95% Vairance Threshold')
plt.title('Explained Variance by PCA Components')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.legend()
plt.grid()
plt.show()
```

```python
pca_optimal = PCA(n_components=9)
pca_transformed = pca_optimal.fit_transform(x_scaled)
```

3. Clustering with Original Data

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

kmeans = KMeans(n_clusters=3, random_state=42)
clusters_original = kmeans.fit_predict(x_scaled)

silhouette = silhouette_score(x_scaled, clusters_original)
print(f'Silhouette Score (Original Data): {silhouette}')
```

```python
plt.scatter(x_scaled[:,0], x_scaled[:,1], c=clusters_original, cmap='viridis', s=50)
plt.title('K-Means Clustering (Original Data)')
plt.show()
```

4. Clustering with PCA Data

```python
kmeans = KMeans(n_clusters=3, random_state=42)
clusters_pca = kmeans.fit_predict(pca_transformed)

silhouette_pca = silhouette_score(pca_transformed, clusters_pca)
print(f'Silhouette Score (PCA Data): {silhouette}')
```

```python
plt.scatter(pca_transformed[:,0], pca_transformed[:,1], c=clusters_pca, cmap='plasma', s=50)
plt.title('K-Means Clustering (PCA Data)')
plt.xlabel('Principal Components 1')
plt.ylabel('Principal Compinents 2')
plt.show()
```

5. Comparison & Analysis

```python
print(f'Silhouette Score Comparison:')
print(f'Original Data: {silhouette}')
print(f'PCA Data: {silhouette_pca}')
```
