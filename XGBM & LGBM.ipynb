2. Data Preprocessing

```python
train['Age'].fillna(train['Age'].median(), inplace=True)
train['Embarked'].fillna(train['Embarked'].mode()[0], inplace=True)
test['Age'].fillna(test['Age'].median(), inplace=True)
test['Fare'].fillna(test['Fare'].median(), inplace=True)
```

```python
train.drop(columns=['Cabin'], inplace=True)
test.drop(columns=['Cabin'], inplace=True)
```

```python
train.isnull().sum()
```

```python
train = pd.get_dummies(train, columns=['Sex','Embarked','Pclass'], drop_first=True)
test = pd.get_dummies(test, columns=['Sex','Embarked','Pclass'], drop_first=True)
```

3. Building Predictive Models

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
```

```python
x = train.drop(columns=['Survived','Name','Ticket','PassengerId'])
y = train['Survived']
```

```python
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
test = test.drop(['PassengerId','Name','Ticket'], axis=1)
test
```

```python
rf = RandomForestRegressor()
rf.fit(x_train, y_train)
y_pred = rf.predict(test)
y_pred
```

```python
df = pd.DataFrame(y_pred, columns=['Survived'])
df = df.astype('int64')
print(df.dtypes)
```

```python
import lightgbm as lgb
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
lgb_model = lgb.LGBMClassifier(random_state=42)
lgb_model.fit(x_train, y_train)
lgb_preds = lgb_model.predict(x_val)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
xgb_model = XGBClassifier(random_state=42, use_label_encoder=False, eval_metrics='logloss')
xgb_model.fit(x_train, y_train)
xgb_preds = xgb_model.predict(x_val)
```

```python
print('LightGBM Classification Report:')
print(classification_report(y_val, lgb_preds))
print('XGBoost Classification Report:')
print(classification_report(y_val, xgb_preds))
print('LightGBM Accuracy:', accuracy_score(y_val, lgb_preds))
print('XGBoost Accuracy:', accuracy_score(y_val, xgb_preds))
```

4. Comparative Analysis

```python
lgb_metrics = classification_report(y_val, lgb_preds, output_dict=True)
xgb_metrics = classification_report(y_val, xgb_preds, output_dict=True)

print('LightGBM F1-Score:', lgb_metrics['weighted avg']['f1-score'])
print('XGBoost F1-Score:', xgb_metrics['weighted avg']['f1-score'])
```

```python
metrics = ['precision', 'recall', 'f1-score']
lgb_scores = [lgb_metrics['weighted avg'][metric] for metric in metrics]
xgb_scores = [xgb_metrics['weighted avg'][metric] for metric in metrics]

plt.bar(metrics, lgb_scores, color='blue', alpha=0.6, label='LightGBM')
plt.bar(metrics, xgb_scores, color='orange', alpha=0.6, label='XGBoost')
plt.title("Comparison of Metrics: LightGBM vs. XGBoost")
plt.ylabel("Score")
plt.legend()
plt.show()
```
