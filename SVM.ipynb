2. Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoded_data = df.copy()
for col in df.select_dtypes(include=['object']).columns:
    encoder = LabelEncoder()
    encoded_data[col] = encoder.fit_transform(df[col])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = encoded_data.drop('class', axis=1)
y = encoded_data['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

3. Data Visualization

```python
sns.pairplot(encoded_data, hue='class', vars=numerical_features)
plt.show()
```

```python
sns.countplot(x='class', data=df)
plt.title('Class Distribution')
plt.show()
```

4. SVM Implementation

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(x_train, y_train)
```

```python
y_pred = svm_model.predict(x_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

5. Visualization of SVM Result

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Greens')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

6. Parameter Tuning & Optimization

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.1,1,10],'kernel':['linear','poly','rbf'],'gamma':['scale','auto']}
```

```python
svm_model = SVC()
grid_search = GridSearchCV(estimator=svm_model, param_grid=param_grid, scoring='accuracy', cv=5, verbose=1)
```

```python
param_grid = {
    'C':[0.1,1,10],
    'kernel':['linear','rbf'],
    'gamma':['scale','auto']
}

grid_search = GridSearchCV(SVC(), param_grid, scoring='accuracy', cv=3, n_jobs=-1)
grid_search.fit(x_train, y_train)
```

```python
print('Best Parameters:', grid_search.best_params_)
print('Best Score:', grid_search.best_score_)
```

7. Comparison & Analysis

```python
kernels = ['linear','poly','rbf','sigmoid']
results = {}

for kernel in kernels:
    model = SVC(kernel=kernel, random_state=42)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = (y_pred == y_test).mean()
    results[kernel] = accuracy

print('Kernel Comaprison:\n', results)

plt.bar(results.keys(), results.values())
plt.title('SVM Kernel Performance Comparison')
plt.ylabel('Accuracy')
plt.show()
```
